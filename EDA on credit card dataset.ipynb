## Handle Missing Data


```python
credit.isnull().sum()  #identifying the missing values
```

```python
credit = credit.dropna(subset=['NPA Status','age','Gender','Region','DebtRatio','NumberOfTimes90DaysLate','Good_Bad'])
```

```python
credit['MonthlyIncome'].fillna(credit['MonthlyIncome'].median(), inplace=True)  #Filled numerical columns with median because it robust the outliers
credit['MonthlyIncome.1'].fillna(credit['MonthlyIncome.1'].median(), inplace=True)
```

```python
credit['NumberOfDependents'] = credit['NumberOfDependents'].fillna(credit['NumberOfDependents'].mode()[0])
```

```python
credit.isnull().sum()                                     #after removing the missing values again checking
```

## Remove Duplicates


```python
credit.duplicated().sum() #checking the duplicates
```

```python
credit=credit.drop_duplicates()
```

```python
credit.duplicated().sum() #Checked and removed duplicate rows to prevent biased analysis.
```

## Fix Inconsistencies in Categorical Data

```python
cat_cols = credit.select_dtypes(include='object').columns
cat_cols
```

```python
credit[cat_cols] = credit[cat_cols].apply(lambda x: x.str.strip().str.lower()) #removing extra spaces and converting to lowercase to ensure consistent category values.
```

## Convert Data Types


```python
credit.dtypes
```

```python
credit['NumberOfDependents'] = pd.to_numeric(credit['NumberOfDependents'], errors='coerce') #coverting columns to appropiate datatypes
```

```python
credit[cat_cols] = credit[cat_cols].astype('category')   
```
